#Theory Questions

**1. What is the difference between a function and a method in Python?**

A function is a block of code that performs a specific task and is defined independently using the `def` keyword. It can be called by its name alone. A method, on the other hand, is a function that is associated with an object and is called using the dot (`.`) operator. While both functions and methods perform operations, methods operate on the data of the object they belong to.

*Example:*  
```python
# Function
def greet(name):
    return f"Hello, {name}"

# Method
text = "hello"
print(text.upper())  # upper() is a method
```

**2. Explain the concept of function arguments and parameters in Python.**

In Python, parameters are the variable names listed in a function's definition, while arguments are the actual values passed to the function when it is called. Parameters define the expected input; arguments provide the actual input during function execution.

*Example:*  
```python
def add(x, y):  # x and y are parameters
    return x + y

add(5, 3)  # 5 and 3 are arguments
```

**3. What are the different ways to define and call a function in Python?**

Functions can be defined using the `def` keyword and called using several types of arguments:

- **Positional arguments**
- **Keyword arguments**
- **Default arguments**
- **Variable-length arguments** using `*args` and `**kwargs`

*Example:*  
```python
def greet(name="User"):
    return f"Hello, {name}"

greet()                # Default argument
greet("Harsh")         # Positional argument
greet(name="Sharma")   # Keyword argument
```

**4. What is the purpose of the `return` statement in a Python function?**

The `return` statement is used to exit a function and send a result back to the caller. It allows a function to provide output, which can be stored or used in further operations.

*Example:*  
```python
def square(n):
    return n ** 2

print(square(4))  # Output: 16
```

**5. What are iterators in Python and how do they differ from iterables?**

An iterable is an object capable of returning its elements one at a time (e.g., list, tuple, string). An iterator is an object that keeps track of its state during iteration and can return the next item using the `next()` function. Iterators implement both `__iter__()` and `__next__()` methods. All iterators are iterables, but not all iterables are iterators.

*Example:*  
```python
my_list = [1, 2, 3]
it = iter(my_list)
print(next(it))  # 1
print(next(it))  # 2
```

**6. Explain the concept of generators in Python and how they are defined.**

Generators are a special type of iterator that yield items one at a time using the `yield` keyword instead of `return`. They are useful when working with large datasets or infinite sequences, as they don’t store the entire result in memory.

*Example:*  
```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for i in countdown(3):
    print(i)
```

---

**7. What are the advantages of using generators over regular functions?**

Generators provide the following advantages:

- **Memory efficiency**: They yield one value at a time instead of storing all items in memory.
- **Lazy evaluation**: Values are produced only when needed.
- **State retention**: Generators automatically resume from where they left off.

*Example:*  
```python
def generate_numbers():
    for i in range(1, 1000000):
        yield i
```

**8. What is a lambda function in Python and when is it typically used?**

A lambda function is an anonymous, inline function defined using the `lambda` keyword. It is typically used when a simple function is needed for a short period, especially within functions like `map()`, `filter()`, or `sorted()`.

*Example:*  
```python
square = lambda x: x * x
print(square(5))  # Output: 25
```

**9. Explain the purpose and usage of the `map()` function in Python.**

The `map()` function applies a specified function to all items in an iterable (like a list) and returns a map object, which can be converted to a list. It is useful for transforming data without writing explicit loops.

*Example:*  
```python
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x ** 2, nums))
print(squares)  # Output: [1, 4, 9, 16]
```

**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

- `map()` applies a function to every item of an iterable and returns the result.
- `filter()` selects items that meet a specific condition.
- `reduce()` (from the `functools` module) applies a rolling computation to sequential elements and returns a single result.

*Example:*  
```python
from functools import reduce

nums = [1, 2, 3, 4]

mapped = list(map(lambda x: x * 2, nums))        # [2, 4, 6, 8]
filtered = list(filter(lambda x: x % 2 == 0, nums))  # [2, 4]
reduced = reduce(lambda x, y: x + y, nums)       # 10
```

````python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

# Example usage:
sample_list = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_of_even_numbers(sample_list))  # Output: 12
````

````python
# 2. Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(text):
    return text[::-1]

# Example usage:
sample_text = "Hello World"
print("Reversed string:", reverse_string(sample_text))  # Output: dlroW olleH
````

````python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_numbers(numbers):
    squared_list = [num ** 2 for num in numbers]
    return squared_list

# Example usage:
sample_list = [1, 2, 3, 4, 5]
print("Squared list:", square_numbers(sample_list))  # Output: [1, 4, 9, 16, 25]
````

````python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime
    for i in range(2, int(n ** 0.5) + 1):  # Check up to square root of n
        if n % i == 0:
            return False
    return True

# Check prime numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is a prime number")
````

````python
# 5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

# Example usage:
for num in fibonacci(10):
    print(num)
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

# Example usage:
for val in powers_of_two(5):
    print(val)
````

````python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()  # .strip() removes extra newline characters

# Example usage:
for line in read_file_line_by_line("sample.txt"):
    print(line)

````

````python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# Sample list of tuples
data = [(3, 5), (1, 2), (4, 1), (2, 4)]

# Sort using lambda on second element (index 1)
sorted_data = sorted(data, key=lambda x: x[1])

print("Sorted list:", sorted_data)
````

````python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# List of temperatures in Celsius
celsius = [0, 20, 37, 100]

# Formula: F = (C × 9/5) + 32
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Temperatures in Fahrenheit:", fahrenheit)
````

````python
# 10. Create a Python program that uses 'filter()' to remove all the vowels from a given string.

def remove_vowels(text):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, text))  #''.join(...) combines the filtered characters back into a string.

# Example usage
input_string = "Hello, World!"
result = remove_vowels(input_string)
print("String without vowels:", result)


````
